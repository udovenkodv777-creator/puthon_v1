# <a id="list">1. Списки и их примеры </a>

Список это та же переменная в python только особого типа list.

Чтобы создать такую переменную, используйте квадратные скобки:

# <a id="while-for">3. Циклы и списки </a>

Самый распространённый сценарий использования списка — это цикл: проходите по элементам списка и что-либо с ними делаете. 

Допустим дан список из целых чисел. Необходимо увеличить каждый элемент списка на 2.

```python
list = [2, 4, 3, 5, 8]

i = 0
while i < len(list):
    list[i] *= 2 #list[i] = list[i] * 2
    i += 1
print(list)
```

Для перебора элементов можно также использовать цикл for и функцию range. Так конструкция будет короче:

```python
for i in range(0, len(list)):
    list[i] *= 2
print(list)
```

```python
list = [2, 4, 3, 5, 8]
```

```python
for i in list:
    i = i * 2
print(list)
```

```python
list_new = [ z for z in range(6)]
print(list_new)

mul_2 = [ x*2 for x in list]
print(mul_2)

```

# <a id="append">4. Добавление элементов в список </a>

Для того, чтобы добавлять элементы в список, можно использовать метод append.Чтобы вызвать метод необходимо написать имя переменной.append(), а в скобках указать значение, которое необходимо добавить в список.

```python
list.append(100)
print(list)
```

Задача: даны оценки студентов группы, рассчитайте среднее значение оценки по курсу.

```python
grades = [5, 5, 4, 3, 4, 5 ,5]
print(grades)
total = 0
for i in grades:
    total += i
print(total/len(grades))
```

```python
grades = [5, 5, 4, 3, 4, 5 ,5]
print(sum(grades)/len(grades))
```

Предположим, что помимо текущей группы, у вас появился 2-ой параллельный поток студентов по этому же курсу. Их результаты записаны в списках grades1, grades2

```python
grades1 = [5, 5, 4, 3, 4, 5, 5]
grades2 = [5, 4, 4, 2, 3, 3, 3, 5, 4, 2]
```

И теперь нам необходимо посчитать среднюю успеваемость по всему потоку сразу, т.е нам надо объединить 2 списка в один.

Вариант 1 (использовать оператор "+" сложения)

```python
all_grades = grades1 + grades2
print(all_grades)
```

```python
print(grades1)
print(grades2)
```

Запомните: с помощью оператора «плюс» можно объединять несколько списков. В результате получится новый объединённый список. \
Его нужно сохранить в отдельную переменную, что мы и сделали: записали всё в переменную all_grades. \
Однако оригинальные списки остались без изменений.

Вариант 2 (использовать метод extend). Вызовем метод extend у нашего первого списка, а в качестве аргумента передадим второй список.

```python
grades1.extend(grades2)
print(grades1)
print(grades2)
```

# <a id="sort">5. Сортировка </a>

Сортировка – одна из самых важных операций над списками. Для этого в python есть специальная функция sorted

```python
#в порядке возрастания(по умолчанию)
print(all_grades)
sorted(all_grades)
```

У функции sorted есть параметр reverse с английского противоположный, если установить что данный параметр равен True, то список отсортируется в обратном порядке

```python
sorted(all_grades, reverse = True)
```

```python
sorted = sorted(all_grades)
print(sorted)
```

```python
list = [x for x in range(1, 6)]
print(list)
```

```python
list.sort(reverse = True)
print(list)
```

```python
example = [ 5, 100, -3, 4, 15]
print(example)
```

```python
example.reverse()
print(example)
```

```python
alph = ['a','b','c','D','E','F']
print(alph)
```

```python
ralf = [ x for x in reversed(alph)]
print(alph)
print(ralf)
```

# <a id="slicing">7. Срезы элементов списка </a>

Вариант 2: Срезы (slice) в Python — это мощный инструмент для работы с последовательностями. С их помощью можно обращаться к отдельным элементам, подпоследовательностям.

Синтаксис срезов : a[start : stop : step]

```python
grades1 = [5, 5, 4, 3, 4, 5, 5]
grades2 = [5, 4, 4, 2, 3, 3, 3, 5, 4, 2]
all_grades = grades1 + grades2
print(all_grades)
```

```python
s2 = [5, 5, 4, 3, 4, 5, 5, 5, 4, 4, 2, 3, 3, 3, 5, 4, 2]
s2
```

```python
s2[14:len(s2)-2]
```

```python
s2[2:-2]
```

```python
s2[2:-2:2]
```

```python
s2[2:]
```

```python
s2[:-2]
```

```python
s2[ : :-1]
```

```python
a = [1, 2, 10, 15, 5, 0]
```

```python
del a[0]
print(a)
```

```python
a.remove(2)
```
