####**Что нужно сделать**
Нужно посчитать частоты каждого числа: сколько раз каждое число встречается в списке `list1`. 

Для этого составьте и запрограммируйте алгоритм, который посчитает частоты всех чисел и выведет на экран результат в формате `число: частота` для каждого числа от `0` до `9`. 

Например, для заданного списка результат должен быть таким:

```bash

1: 5
2: 1
3: 2
4: 1
5: 3
6: 1

9: 1
```

Убедитесь, что алгоритм работает правильно для разных входных данных. Создайте несколько разных списков и проверьте алгоритм на них.


#### **Подсказка**

Для подсчёта частот используйте дополнительный список, индексами в котором будут значения `list1` (по условию задачи они всегда находятся в диапазоне от `0` до `9`). 

````python
list1 = [1, 3, 7, 1, 1, 2, 3, 7, 6, 5, 5, 4, 1, 5, 9, 1, 7, 0, 0]

# Ваш код здесь

print(list1)

for i in range(10):
    count = list1.count(i)
    print(i,' посторяется столько раз: ', count)
````

## Задача 2. Поиск единственного числа

Дан список `list2`. Элементы этого списка – целые неотрицательные числа от 0 до 100. Список составлен так, что каждое число в нём (кроме одного) встречается дважды. 

**Что нужно сделать** 
Составьте и запрограммируйте алгоритм, который находит в списке `list2` то единственное число, которое встречается в нём один раз. 

Решите задачу как минимум двумя способами (смотрите подсказку).

Убедитесь, что алгоритм работает правильно для разных входных данных. Создайте несколько разных списков и проверьте алгоритм на них.

__Подсказка__

Задачу можно решить по-разному. 

**Способ №1**. Наиболее очевидный способ – взять первый элемент `list2` (с индексом нуль), и, пробежавшись по списку, найти его пару. Если не находится, то первый элемент и есть искомое одинокое число. Если же пара нашлась, то нужно взять второй элемент и повторить всё сначала. И так до тех пор, пока не найдётся одинокое число. Реализуйте этот алгоритм.

**Способ №2**. Применить трюк, который использовался в предыдущей задаче: посчитать частоты слов, используя другой массив (важно учитывать, что числа в `list2` уже от 0 до 100) - `freqs`. Потом пробежаться по `freqs` и найти элемент, который равен единице. Позиция этого элемента и будет искомым одиноким числом. Реализуйте и этот алгоритм тоже. Чтобы определить список из ста нулей, можно использовать вот такую конструкцию: `freqs = [0]*100`. 

````python
# Способ 1
list2 = [0, 1, 3, 4, 7, 98, 3, 777, 98, 7, 16, 16, 4, 0, 1]

# Ваш код
for x in list2:
    if list2.count(x) == 1:
        print(f"Единственное число: {x}")
        break 
````

````python
# Способ 2

# Ваш код
# 1. Создаем список-счетчик из 101 нуля (так как числа до 100 включительно)
freqs = [0] * 101

# Допустим, у нас есть такой список list2 (пример)
list2 = [0, 1, 3, 4, 7, 98, 3,15, 98, 7, 16, 16, 4, 0, 1]

# 2. Считаем частоты: число из list2 становится индексом в freqs
for x in list2:
    freqs[x] += 1

# 3. Ищем индекс элемента, значение которого равно 1
for i in range(len(freqs)):
    if freqs[i] == 1:
        print(f"Одинокое число: {i}")
        break
````

### Задача 3. Сортировка и слияние файлов

Представьте себе, что вы работаете на компьютере с очень ограниченной памятью. На жёстком диске этого компьютера лежат два больших файла `file1` и `file2`, в каждом файле – числа. Считать в память компьютера можно только один файл, но никак не оба одновременно. **Необходимо записать на жёсткий диск третий файл, который будет содержать числа из обоих файлов, отсортированные в порядке возрастания.**  

Пусть они выглядят так:

|file1 |file2 |
|------|------|
|  1   | 2    |
|  5   | 3    |
|  6   | 5    |
|  18  | 7    |
|  42  | 8    |
|  74  | 18   |
|  80  | 19   |
|  87  | 20   |
|  91  | 43   |
|  99  | 50   |
|  100 | 81   |
|      | 93   |
|      | 101  |
|      | 102  |
|      | 103  |

По отдельности каждый файл отсортирован, но вам нужно объединить эти два файла (или, как еще говорят _слить_) в один отсортированный файл. Считать их в память, соединить и отсортировать не получится, потому что вдвоем одновременно они в память не поместятся. Поэтому нужен определённый алгоритм, который позволит записывать объединённый отсортированный файл, не считывая в память два оригинальных файла. Формально такой алгоритм называется [сортировка слиянием](https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC#:~:text=%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0%20%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC%20(%D0%B0%D0%BD%D0%B3%D0%BB.,%D0%BF%D1%80%D0%B8%D0%BD%D1%86%D0%B8%D0%BF%D0%B0%20%C2%AB%D1%80%D0%B0%D0%B7%D0%B4%D0%B5%D0%BB%D1%8F%D0%B9%20%D0%B8%20%D0%B2%D0%BB%D0%B0%D1%81%D1%82%D0%B2%D1%83%D0%B9%C2%BB.). 

Вам потребуется только часть этого алгоритма, потому что исходные файлы уже отсортированы. Остаётся только слить их в единый файл, который тоже будет отсортирован. Разберем алгоритм `слияния` по шагам:

1. Берём `первые` элементы каждого файла и сравниваем их между собой.
1. `1` меньше, чем `2`, поэтому сохраняем `1` в файл `result` (будет [1]).
1. Берём `второй` элемент из `file1` и сравниваем его с `первым элементом` из `file2`.
1. `5` больше, чем `2`, поэтому сохраняем `2` в файл `result` (будет [1, 2]).
1. Берём `второй` элемент из `file2` и сравниваем с ним `второй` элемент из `file1`.
1. `5` больше, чем `3`, поэтому сохраняем `3` в файл `result` (будет [1, 2, 3]).
1. Берём `третий` элемент из `file2` и сравниваем с ним `второй` элемент из `file1`.
1. `5` равно `5`, поэтому сохраняем `5` в файл `result` (будет [1, 2, 3, 5]). 
1. Можно взять `третий` элемент `file1` и сравнить его с `третьим` элементом `file2` или взять `четвёртый` элемент `file2` и сравнить с ним `второй` элемент из `file1`. Пусть будет последний вариант.
1. `5` меньше, чем `7`, поэтому сохраняем `5` в файл `result` (будет [1, 2, 3, 5, 5]).
1. Берём `третий` элемент из `file1` и сравниваем его с `четвёртым` элементом из `file2`.
1. `6` меньше, чем `7`, поэтому сохраняем `6` в файл `result` (будет [1, 2, 3, 5, 5, 6]).
1. Берём `четвёртый` элемент из `file1` и сравниваем его с `четвёртым` элементом из `file2`.
1. Продолжаем в этом духе, пока не дойдём до конца одного из файлов.

Когда дойдём до конца одного из файлов, в файле `result` будет отсортированная `смесь` элементов `file1` и `file2`, а в одном из файлов (до конца которого вы не дошли) могут быть ещё элементы, которые гарантированно не меньше элементов в `result`. Вам остаётся `приклеить` эти элементы в конец `result`, и тогда в ` result` окажется отсортированное объединение `file1` и `file2`. Задача решена.  

**Формальная постановка задачи**
