5* Изменить данные в старом списке

```python
salary = [35000, 45000, 35000, 44000, 55000]

# Ваш код
# Создаем новый список, увеличивая каждое число на 10%
new_salary = [int(x * 1.1) for x in salary]

print(new_salary)
```

```python
salary1 = [35000, 45000, 35000, 44000, 55000]
new_salary1 = []

for s in salary1:
    increased = round(s * 1.1)
    new_salary1.append(increased)

print(new_salary1)
```

```python
#Вывод результата в виде красивой таблицы
salary = [35000, 45000, 35000, 44000, 55000]
new_salary = [int(s * 1.1) for s in salary]

print(f"{'№':<5} | {'Старая ЗП':<15} | {'Новая ЗП':<15}")
print("-" * 40)

for i, (old, new) in enumerate(zip(salary, new_salary), 1):
    print(f"{i:<5} | {old:<15} | {new:<15}")
```

```python
from tabulate import tabulate

salary = [35000, 45000, 35000, 44000, 55000]
data = [[i+1, s, int(s*1.1)] for i, s in enumerate(salary)]

# Печать таблицы в стиле "grid"
print(tabulate(data, headers=["№", "Старая ЗП", "Новая ЗП (+10%)"], tablefmt="grid"))
```

```python
import pandas as pd

df = pd.DataFrame({
    'Старая ЗП': salary,
    'Новая ЗП': [int(s * 1.1) for s in salary]
})

print(df)
```
